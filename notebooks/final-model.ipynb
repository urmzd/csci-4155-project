```python
from src.bootstrap import bootstrap, TrainerFactory
from src.data.data_handler import BaseModel
from pathlib import Path
from pytorch_lightning.loggers import WandbLogger
import torch

trainer_factory = TrainerFactory(
    strategy=None, 
    max_epochs=50, 
)

# K Fold -- check
# batch size -- check
# drop out -- 
# pooling
# convolutions
# depth 
# width
# optimizers

class FinalModel(BaseModel):

  def __init__(self, momentum: float = 0.99, n_targets=2, learning_rate=1e-5, dropout_p=0.5):
    self.momentum = momentum
    self.dropout_p = dropout_p

    super().__init__(n_targets=n_targets, learning_rate=learning_rate)

    self.layers = torch.nn.Sequential(
      torch.nn.Conv2d(1, 16, 3),
      torch.nn.AvgPool2d(2, 2),
      torch.nn.ReLU(),
      torch.nn.Conv2d(16, 32, 3),
      torch.nn.AvgPool2d(2, 2),
      torch.nn.ReLU(),
      torch.nn.Conv2d(32, 64, 3),
      torch.nn.AvgPool2d(2, 2),
      torch.nn.ReLU(),
      torch.nn.Dropout2d(dropout_p),
      torch.nn.Flatten(1, -1),
      torch.nn.Linear(64 * 23 ** 2, n_targets),
      torch.nn.LogSoftmax()
    )

  def forward(self, X):
    return self.layers(X)

  def configure_optimizers(self):
    return torch.optim.SGD(self.parameters(), lr=self.learning_rate, momentum=self.momentum)

bootstrap(
    trainer_factory=trainer_factory,
    data_manifest_path=Path("/content/lepus-classifier/resources/data.csv"),
    model=FinalModel(),
    batch_size=2,
    num_folds=3,
)
```
